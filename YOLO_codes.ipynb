```python
import torch
import cv2
from ultralytics import YOLO

# Set device to CPU explicitly
device = 'cpu'

# Load the pre-trained YOLOv8 model (YOLOv8n for smaller, faster model)
model = YOLO('yolov8n.pt').to(device)

# Function to run YOLO object detection on video input
def detect_objects_in_video(video_path):
    # Open video file or capture device (0 for webcam)
    cap = cv2.VideoCapture(video_path)

    # Check if video file or webcam opened successfully
    if not cap.isOpened():
        print("Error: Could not open video source.")
        return

    # Get video frame width and height
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))

    # Define the codec and create VideoWriter object to save output
    out = cv2.VideoWriter('output_video.mp4', cv2.VideoWriter_fourcc(*'MP4V'), 10, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()  # Read frame-by-frame
        if not ret:
            break  # Break loop when no more frames are available

        # Run object detection using YOLO (on CPU)
        results = model(frame, device='cpu')  # Model processes the frame on CPU

        # Parse YOLO results and draw bounding boxes on the frame
        annotated_frame = results[0].plot()  # Get annotated frame with detections

        # Display the frame with detections
        cv2.imshow('YOLO Real-Time Object Detection', annotated_frame)

        # Write the frame to the output video file
        out.write(annotated_frame)

        # Break loop on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the video capture and writer
    cap.release()
    out.release()
    cv2.destroyAllWindows()

# Detect objects on a sample video (replace '0' with file path for video input)
detect_objects_in_video(0)  # 0 opens the webcam for real-time detection
```

```python
import torch
import cv2
import pandas as pd

# Load the YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Function to process video and extract detection information frame by frame
def process_video_and_extract_detections(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    # List to store detection data
    detections = []

    # Initialize frame count
    frame_count = 0

    # Process each frame
    while cap.isOpened():
        ret, frame = cap.read()  # Read frame-by-frame
        if not ret:
            break

        frame_count += 1

        # Perform object detection on the frame
        results = model(frame)

        # Extract the relevant data (bounding boxes, labels, confidence scores)
        for detection in results.xyxy[0]:  # Results for each frame in [xmin, ymin, xmax, ymax, confidence, class]
            xmin, ymin, xmax, ymax, confidence, class_id = detection.tolist()
            label = results.names[int(class_id)]  # Convert class_id to label

            # Append detection data for this frame
            detections.append({
                'frame': frame_count,
                'xmin': int(xmin),
                'ymin': int(ymin),
                'xmax': int(xmax),
                'ymax': int(ymax),
                'confidence': float(confidence),
                'class_id': int(class_id),
                'label': label
            })

    # Release the video capture
    cap.release()

    # Convert detections to a DataFrame for easier handling and later use
    detections_df = pd.DataFrame(detections)

    # Save detection data to a CSV file for later analysis
    detections_df.to_csv('detections.csv', index=False)

    return detections_df

# Run the detection and save results
video_path = '/content/drive/MyDrive/DL/The two talking cats.mp4'  # Replace with actual video path
detections_df = process_video_and_extract_detections(video_path)

# Show the first few rows of the detection data
detections_df.head()
```

```python
import pandas as pd

# Assuming detections_df is your original DataFrame from YOLO output
# Example function to preprocess the YOLO output
def preprocess_detections_to_excel(detections_df, frame_width, frame_height, output_filename='preprocessed_detections.xlsx'):
    # Initialize a new DataFrame for preprocessed detections
    preprocessed_data = []

    # Initialize a dictionary to track object IDs
    object_ids = {}
    next_object_id = 0

    # Process each detection
    for _, row in detections_df.iterrows():
        frame = row['frame']
        class_id = row['class_id']

        # Calculate normalized coordinates
        normalized_xmin = row['xmin'] / frame_width
        normalized_ymin = row['ymin'] / frame_height
        normalized_xmax = row['xmax'] / frame_width
        normalized_ymax = row['ymax'] / frame_height

        # Create a unique key for the object based on frame and class_id
        object_key = (frame, class_id)

        # Assign object ID if it's a new object
        if object_key not in object_ids:
            object_ids[object_key] = next_object_id
            next_object_id += 1

        # Store the preprocessed detection
        preprocessed_data.append({
            'frame': frame,
            'object_id': object_ids[object_key],
            'normalized_xmin': normalized_xmin,
            'normalized_ymin': normalized_ymin,
            'normalized_xmax': normalized_xmax,
            'normalized_ymax': normalized_ymax,
            'confidence': row['confidence'],
            'class_id': class_id,
            'label': row['label']
        })

    # Convert to DataFrame
    preprocessed_df = pd.DataFrame(preprocessed_data)

    # Save the DataFrame to an Excel file
    preprocessed_df.to_excel(output_filename, index=False)

    return preprocessed_df

# Example usage
frame_width = 640  # Replace with actual frame width
frame_height = 480  # Replace with actual frame height
preprocessed_df = preprocess_detections_to_excel(detections_df, frame_width, frame_height)

# Display the first few rows of the preprocessed DataFrame
print(preprocessed_df.head())
```

```python
import cv2
import pandas as pd
import numpy as np
from filterpy.kalman import KalmanFilter
import torch  # Make sure to import torch for YOLOv5

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # Load the YOLOv5 model (small version)

output_video_path = 'output_video2.mp4'
kalman_data = []  # List to store Kalman filter results for each frame

# Function to create a Kalman Filter
def create_kalman_filter():
    kf = KalmanFilter(dim_x=4, dim_z=2)
    kf.x = np.array([[0], [0], [0], [0]])  # Initial state (x, y, dx, dy)
    kf.P *= 1000.  # Initial uncertainty
    kf.F = np.array([[1, 0, 1, 0],
                     [0, 1, 0, 1],
                     [0, 0, 1, 0],
                     [0, 0, 0, 1]])  # State transition matrix
    kf.H = np.array([[1, 0, 0, 0],
                     [0, 1, 0, 0]])  # Measurement function
    kf.R = np.array([[10, 0],
                     [0, 10]])  # Measurement noise
    return kf

# Function to track objects
def track_objects(detections, trackers, frame, frame_id):
    for det in detections:
        bbox = det['bbox']  # [xmin, ymin, xmax, ymax]
        center = np.array([(bbox[0] + bbox[2]) / 2, (bbox[1] + bbox[3]) / 2])  # Center of the bounding box

        # Create or update the Kalman filter for each object
        if det['id'] not in trackers:
            trackers[det['id']] = create_kalman_filter()

        kf = trackers[det['id']]

        # Update Kalman filter with detected center
        kf.predict()
        kf.update(center)

        # Get predicted position
        predicted_pos = kf.x[:2].flatten()

        # Append the Kalman filter results (frame, object id, predicted position) to the data list
        kalman_data.append({
            'Frame': frame_id,
            'Object_ID': det['id'],
            'Predicted_X': predicted_pos[0],
            'Predicted_Y': predicted_pos[1],
            'BBox_Xmin': bbox[0],
            'BBox_Ymin': bbox[1],
            'BBox_Xmax': bbox[2],
            'BBox_Ymax': bbox[3],
        })

        # Draw the predicted position on the frame (for visualization)
        cv2.circle(frame, (int(predicted_pos[0]), int(predicted_pos[1])), 5, (255, 0, 0), -1)

        # Draw bounding box for detected object
        cv2.rectangle(frame, (int(bbox[0]), int(bbox[1])), (int(bbox[2]), int(bbox[3])), (0, 255, 0), 2)
        cv2.putText(frame, f'ID: {det["id"]}', (int(bbox[0]), int(bbox[1]-10)), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    return trackers

# Read input video
cap = cv2.VideoCapture('/content/drive/MyDrive/DL/The two talking cats.mp4')  # Replace with your input video path
fps = cap.get(cv2.CAP_PROP_FPS)
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

# Video writer to save the output
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

trackers = {}  # To store Kalman filters for each detected object
frame_id = 0

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Detect objects using YOLOv5
    results = model(frame)  # Run YOLOv5 inference
    detections = []

    for result in results.xyxy[0]:  # Loop through detections
        xmin, ymin, xmax, ymax, conf, class_id = result
        detections.append({
            'bbox': [xmin.item(), ymin.item(), xmax.item(), ymax.item()],
            'confidence': conf.item(),
            'class_id': int(class_id.item()),
            'id': int(class_id.item())  # Using class ID as a placeholder for object ID
        })

    # Track objects
    trackers = track_objects(detections, trackers, frame, frame_id)

    # Write the frame to the output video
    out.write(frame)

    frame_id += 1  # Increment frame ID

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Convert the Kalman filter results to a pandas DataFrame
kalman_df = pd.DataFrame(kalman_data)

# Save the DataFrame to an Excel file
kalman_df.to_excel('kalman_filter_results.xlsx', index=False)

print(f"Output video saved to: {output_video_path}")
print(f"Kalman filter results saved to: kalman_filter_results.xlsx")
```
